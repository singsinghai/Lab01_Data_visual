## Thêm một số cột mới

### Thêm các cột về tỷ lệ

```python
df['Death rate'] = df['Total Deaths'] / df['Total Cases']
df['Recovery rate'] = df['Total Recovered'] / df['Total Cases']
df['Active rate'] = df['Active Cases'] / df['Total Cases']
df['Serious rate'] = df['Serious, Critical'] / df['Total Cases']
df['Cases per test'] = df['Total Cases'] / df['Total Tests']
```

# Trực quan hóa dữ liệu

## Trực quan hóa trên một trường đơn

Ta có 2 nhóm trường dữ liệu đơn:
- Liên quan đến tổng: Total Cases, Total Tests, Total Recovered, Total Deaths và Active Cases, ... 
- Liên quan đến tỷ lệ: Recovery Rate, Death Rate, Active Rate, ...

Để thể hiện các tính chất của mỗi nhóm trường, nhóm đề xuất phương án như sau: 
- Với các trường liên quan đến tổng: Sử dụng biểu đồ Histogram
- Với các trường liên quan đến tỷ lệ: Sử dụng biểu đồ Box plot

### Histogram

#### `Total Cases` và `Total Test`

Biểu đồ dưới đây thể hiện sự phân bố về tổng số ca mắc và tổng số lượt test của các quốc gia trên thế giới sử dụng histogram. Có thể thấy phân bố của cả hai thuộc tính đều bị lệch dương khá nhiều, trong đó cả hai đều có những trường hợp có giá trị rất lớn.

```python
f = sns.FacetGrid(
    # melt để đưa dữ dữ liệu từ dạng wide-form về dạng long-form
    data=df[['Total Cases', 'Total Tests']].melt(),
    col='variable',
    sharex=False,
    height=5,
)

f.map_dataframe(sns.histplot, x='value')
f.fig.subplots_adjust(top=0.8)
f.fig.suptitle(
    t='Phân bố về tổng số ca mắc và tổng số lượt test của các quốc gia trên thế giới',
    fontsize=14,
    fontweight='bold'
)
plt.show()
```

#### `Total Recovered`, `Total Deaths` và `Active Cases`

Biểu đồ histogram dưới đây thể hiện sự phân bố về tổng ca hồi phục, ca tử vong và số ca còn điều trị của các quốc gia trên thế giới. Một lần nữa ta cũng thấy cả ba thuộc tính đều có phân bố lệch dương, các thuộc tính hầu hết tập trung chủ yếu ở các giá trị nhỏ.

```python
f = sns.FacetGrid(
    # melt để đưa dữ dữ liệu từ dạng wide-form về dạng long-form
    data=df[['Total Recovered', 'Total Deaths', 'Active Cases']].melt(),
    col='variable',
    sharex=False,
    height=5,
)

f.map_dataframe(sns.histplot, x='value')
f.fig.subplots_adjust(top=0.8)
f.fig.suptitle(
    t='Phân bố về tổng số ca hồi phục, tử vong và còn điều trị của các quốc gia trên thế giới',
    fontsize=14,
    fontweight='bold'
)
plt.show()
```

### Box plot

#### `Recovery rate`, `Death rate` và `Active rate`

Biểu đồ dưới đây thể hiện sự phân bố về tỷ lệ hồi phục, tỷ lệ tử vong, tỷ lệ các ca còn đang mắc của các quốc gia trên thế giới. Ở đây ta dùng biểu đồ box plot vì ta đang khảo sát về phân bố của các trường có kiểu dữ liệu số. Hơn nữa, vì cả ba thuộc tính đều có cùng đơn vị là phần trăm, và `Recovery rate` + `Death rate` + `Active rate` = 100% nên ta trực quan bằng ba biểu đồ box plot có cùng scale ở trục x.

Nhìn chung, hầu hết các quốc gia trên thế giới đều đã kiểm soát tốt các ca mắc của nước mình: phần lớn các quốc gia đều có tỷ lệ hồi phục ở mức cao quanh 90-100%, tỷ lệ các ca còn điều trị ở quanh mức dưới 10%. Tỷ lệ tử vong là thấp nhất trong số các tỷ lệ, trung bình 1.4% khả năng tử vong trên mỗi ca mắc (xem boxplot phía dưới để thấy rõ hơn).

## Trực quan và phân tích mối quan hệ giữa nhiều trường dữ liệu

### Correlation Matrix giữa các thuộc tính 


Ta xây dựng correlation matrix và vẽ heatmap để trực quan mối quan hệ giữa các cặp thuộc tính numeric.

```python
corr = df.corr()

def set_value_for_corr(x):
    if x > 0.7:
        return 1
    elif x < -0.7:
        return -1
    else:
        return 0
    

corr = corr.applymap(set_value_for_corr)
plt.figure(figsize=(13, 7))
sns.heatmap(corr, annot=True)
```

Ta có mối quan hệ tính toán sao: Total Cases = Active Cases + Total Deaths + Total Recovered. Vì vậy sau khi tính tỉ lệ Active rate và Recovery rate có mức tương quan nghịch cao.

### Scatterplot + Regression

Vẽ đường hồi quy tuyến tính để xem xét một số mối quan hệ nhân quả giữa các trường dữ liệu

```python
def scatter_with_regression_plot(x_col, y_col):
    fig, ax = plt.subplots()
    sns.scatterplot(x=x_col, y=y_col)
    sns.regplot(x=x_col, y=y_col, ci=None, scatter=False, color='orange', line_kws={'linewidth':3});
```

####  Mối quan hệ giữa các cột `Total Tests`, `Population`, `Total Deaths` và `Total Cases`

- Ý nghĩa của cột `Tests/1M pop` chính là tỉ lệ test trên 1 triệu dân, cũng là thương của cột `Total Tests` và cột `Population`.
- Cột `Death Rate` được tính bằng thương của cột `Total Deaths` và `Total Cases`.

Như vậy qua việc trực quan giữa cột `Test/1M pop` và `Death Rate`, ta sẽ xem được mối quan hệ giữa 4 cột `Total Tests`, `Population`, `Total Deaths` và `Total Cases`.

```python
scatter_with_regression_plot(np.log(df['Tests/1M pop']), np.log(df['Death rate']))
```

Có vẻ như khi tích cực thực hiện test covid, chúng ta sẽ nhanh chóng phòng ngừa được dịch bệnh và giảm tỉ lệ tử vong của những người mắc bệnh.

```python
scatter_with_regression_plot(np.log(df['Total Cases']), np.log(df['Total Deaths']))
```

#### Mối quan hệ giữa tỉ lệ mắc bệnh và tỉ lệ tử vong của 1 quốc gia

Ta sẽ xem xét liệu có mối quan hệ giữa 2 cột `Total Cases/1M pop` và `Deaths/1M pop`. 

```python
scatter_with_regression_plot(np.log(df['Total Cases/1M pop']), np.log(df['Deaths/1M pop']))
```

Có vẻ như một quốc gia có tỉ lệ mắc trên tổng dân số càng cao thì tử lệ tử vong trên tổng dân số cũng cao. Ta tiếp tục trực quan trên 2 cột `Total Cases` và `Total Deaths`.

```python
scatter_with_regression_plot(np.log(df['Total Cases']), np.log(df['Total Deaths']))
```

### Bar plot

- Sử dụng để so sánh tỉ lệ số ca hiện tại giữa các châu lục
- Ta tiến hành nhóm các dòng dữ liệu theo châu lục và tính mean để có các tỉ lệ trung bình theo từng châu lục. Sau đó ta xem xét các chây lục có tỉ lệ số ca đang mắc bệnh và tỉ lệ tử vong giảm dần

```python
grouped_by_continent = df.groupby(['Continent'])
active_cases_mean = grouped_by_continent['Active Cases/1M pop'].mean().sort_values()
death_cases_mean = grouped_by_continent['Deaths/1M pop'].mean().sort_values()
fig, ax = plt.subplots(1, 2, figsize=(17,6))

ax[0].barh(active_cases_mean.keys(), active_cases_mean.values)
ax[0].set_xlabel('Active Cases/1M pop', fontsize=14)
ax[0].set_ylabel('Continent', fontsize=14)
ax[0].set_title('Tỉ lệ số ca đang mắc bệnh giữa từng châu lục', fontsize=20)

ax[1].barh(death_cases_mean.keys(), death_cases_mean.values)
ax[1].set_xlabel('Deaths/1M pop', fontsize=14)
ax[1].set_ylabel('Continent', fontsize=14)
ax[1].set_title('Tỉ lệ số ca tử vong giữa từng châu lục', fontsize=20)
ax[1].yaxis.tick_right()
```

Đi sâu hơn vào từng châu lục, ta xem top 3 nước thuộc châu lục đó có tỉ lệ active case và tỉ lệ tử vong cao nhất

```python
continents = df['Continent'].unique()
for i in range(len(continents)):
    
    fig, ax = plt.subplots(1, 2, figsize=(15, 4))
    fig.suptitle(f'Top những nước có tỉ lệ người đang nhiễm bệnh và tỉ lệ tử vong cao nhất trong {continents[i]}', fontsize=17)
    by_continent = df[df['Continent'] == continents[i]]
    temp = by_continent[['Country', 'Active Cases/1M pop']].set_index('Country').sort_values('Active Cases/1M pop', ascending=False)
    ax[0].barh(temp[:3].index, temp[:3].values.reshape(-1))
    ax[0].set_xlabel('Active Cases/1M pop', fontsize=13)
    
    temp = by_continent[['Country', 'Deaths/1M pop']].set_index('Country').sort_values('Deaths/1M pop', ascending=False)
    ax[1].yaxis.tick_right()
    ax[1].barh(temp[:3].index, temp[:3].values.reshape(-1))
    ax[1].set_xlabel('Deaths/1M pop', fontsize=13)
```

### Mở rộng: Treemap

- Sau khi đã có Bar Chart thể hiện top những nước có tỉ lệ người đang nhiễm bệnh và tỉ lệ tử vong cao nhất trong từng khu vực, nhóm đã nảy ra ý tưởng gom nhóm các quốc gia trong châu lục lại với nhau thành một nhóm, và gom các nhóm trên thành một nhóm lớn hơn. Từ ý tưởng trên, nhóm quyết định sử dụng Treemap
- Biểu đồ Treemap có khả năng thể hiện được so sánh giữa các nhóm giá trị với nhau, cụ thể ở đây là giữa các châu lục trên thế giới, và giữa các quốc gia trong mỗi châu lục.
- Nhóm sử dụng Treemap với các trường `Deaths/1M pop`, và so sánh tổng quan theo nhóm giữa các châu lục với nhau và giữa các nước trong châu lục với nhau.

```python
fig = px.treemap(df, 
                path = [px.Constant("all"),'Continent', 'Country'],
                values = 'Deaths/1M pop',
                hover_data = ['Country Code'],
                color_continuous_scale = px.colors.sequential.Burg,
                title = 'Tỉ lệ tử vong của các quốc gia và châu lục'
                )
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

## Trực quan và phân tích dựa trên vị trí địa lý

- Ta sử dụng Bubble Map để có thể thấy rõ được tình hình tổng thể của các quốc gia trên thế giới với nhau, cũng như điều chỉnh góc nhìn hợp lý cho bản thân để có cái nhìn tổng thể về nhóm các quốc gia trong cùng 1 khu vực, 1 châu lục.

    - Với trường liên quan đến tổng, nhóm sẽ trực quan trường `Total Cases`.
