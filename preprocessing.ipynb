```python
import datetime as dt
import numpy as np
import pandas as pd

pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv('covid.csv', index_col=0)
df
```

# Tiền xử lý

```python
df.dtypes
```

## Chuyển các dữ liệu ở dạng string về dạng số

```python
excludes = ['Country', 'Continent', 'Date', 'New Deaths', '1 Test every X ppl', 'New Deaths/1M pop']
for col in df.columns[~df.columns.isin(excludes)]:
    df[col] = df[col].fillna('')
    df[col] = df[col].apply(lambda x: x.strip().replace(',', ''))
    df[col] = df[col].apply(lambda x: x.strip().replace('+', ''))
    df[col] = df[col].replace({'': np.nan})
    df[col] = pd.to_numeric(df[col])
df
```

```python
df.dtypes
```

## Chuyển cột `Date` về dạng `datetime`

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Loại dòng có tỷ lệ thiếu lớn hơn 40%

```python
df = df[df.isna().sum(axis=1) / len(df.columns) < 0.4]
```
