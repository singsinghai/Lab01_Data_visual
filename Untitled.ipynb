```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('covid.xlsx', index_col='#')
new_df = data.loc[:, ~data.columns.isin(['New Cases', 'New Deaths', 'New Recovered'])]
new_df[new_df.isna().any(axis=1)]
new_df = new_df[new_df.isna().sum(axis=1) / len( new_df.columns) < 0.4]
new_df
```

```python
plt.hist(new_df['Total Cases'] / new_df['Total Recovered'], bins=100);
```

```python
corr = new_df.corr()
plt.figure(figsize=(13,6))
sns.heatmap(corr, annot=True)
```
